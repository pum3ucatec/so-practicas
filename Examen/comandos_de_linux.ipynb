#Guía de Comandos Básicos de Linux

Linux ofrece una gran variedad de comandos útiles para realizar tareas desde la línea de comandos. Aquí tienes una lista con los comandos más utilizados, su descripción y ejemplos prácticos.
#Comandos de Gestión de Archivos y Directorios
#1. ls

Lista el contenido de un directorio.

ls

    Opciones útiles:
        -l: Lista con detalles.
        -a: Muestra archivos ocultos.

Ejemplo:

ls -la

#2. cd

Cambia el directorio actual.

cd [ruta]

Ejemplos:

cd /home/usuario
cd ..
cd ~

#3. pwd

Muestra la ruta del directorio actual.

pwd

Ejemplo de salida:

/home/usuario/proyectos

#4. mkdir

Crea un nuevo directorio.

mkdir [nombre_directorio]

Ejemplo:

mkdir nueva_carpeta



#5. cp

Copia archivos o directorios.

cp [origen] [destino]

    Opciones útiles:
        -r: Copia directorios de forma recursiva.

Ejemplo:

cp archivo.txt /home/usuario/carpeta/
cp -r carpeta1 carpeta2

Comandos de Administración del Sistema
#6 chmod

Cambia los permisos de archivos o directorios.

chmod [permisos] [archivo]

Ejemplo:

chmod 755 script.sh

#7. sudo

Ejecuta comandos con privilegios de superusuario.

sudo [comando]

Ejemplo:

sudo apt update

#8. git fetch
descarga las actualizaciones del repositorio remoto
Este comando descarga los cambios del repositorio remoto origin, pero no modifica tu rama actual.
#9. git pull
Es una combinación de git fetch seguido de git merge. Es decir, descarga los cambios del repositorio remoto y fusiona automáticamente
